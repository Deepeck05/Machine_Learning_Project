NumPy numeric pyhthon


```python
import numpy as np
```

```python
# creating Array and matrix
a = np.array([1, 2, 3])
b = np.array([[1, 2, 3], [4, 5, 6]])

print(a)
print(b)
```

```python
#Arithmetic
x = np.array([10, 20, 30])
y = np.array([1, 2, 3])

print(x + y)
print(x - y)
print(x * y)
print(x / y)
```

```python
# Ststistic
data = np.array([10, 20, 30, 40, 50])

print(np.mean(data))
print(np.median(data))
print(np.std(data))
print(np.sum(data))
print(np.max(data))
print(np.min(data))
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.reshape(3, 2))  # Reshape into 3 rows, 2 cols
print(arr.flatten())   #Matrix array in flat
```

```python
a = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print(np.dot(a, B))  # Matrix multiplication
print(B.T)   #The column will transpose a row
```

```python
arr = np.array([[10, 20, 30],
                [40, 50, 60]])

print(arr[1,2])     # Single element → 20
print(arr[1])        # Entire 2nd row → [40 50 60]
print(arr[:,1])     # Entire 2nd column → [20 50]
print(arr[1, 0:2])  # 2nd row, 1st and 2nd columns → [40 50]
```

```python
print(np.zeros((2,3)))  # 2x3 matrix of zeros
print(np.ones((2,2)))   # 2x2 matrix of ones
print(np.eye(3))        # 3x3 Identity matrix
```

Comparision Operator in Array

```python
arr = np.array([10, 20, 30, 40])
print(arr > 25)      # [False False  True  True]
print(arr[arr > 25]) # Filter values → [30 40]
```
